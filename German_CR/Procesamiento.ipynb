## Procesamiento de datos a categorÃ­as

```python
df_2 = pd.read_csv('../Datasets/scored_data_no_null.csv', index_col=0)
df_2.head()
```

# Creamos categorias a intervalos 
## Edades

```python
intervalo_edad = ( 0,25, 35, 60, 120)

categorias_edad = ['Estudiante', 'Joven Adulto', 'Adulto', 'Mayor edad']
df_2['age'] = pd.cut(df_2.age, intervalo_edad, labels = categorias_edad)

df_2
```

# Binario 

```python
df_2 = df_2.merge(pd.get_dummies(df_2.age, prefix='age'), 
                  left_index=True, right_index=True)

df_2 = df_2.drop(columns = 'age')
df_2
```

## duration

```python
# Categorias 
intervalo_aÃ±os = (0, 12, 24, 36, 48, 60, 72)
categorias_aÃ±os =['1 aÃ±o', '2 aÃ±os', '3 aÃ±os', '4 aÃ±os', '5 asÃ±os', '6 aÃ±os']
df_2['duration'] = pd.cut(df_2.duration, intervalo_aÃ±os, labels= categorias_aÃ±os)

# Binario 
df_2 = df_2.merge(pd.get_dummies(df_2.duration, prefix='duration'), 
                  left_index=True, right_index=True)

df_2 = df_2.drop(columns = 'duration')
df_2
```
